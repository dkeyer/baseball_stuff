# Scraping Individual Player Stats

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
from __future__ import print_function

from datetime import date
import datetime
#imports
```

```python
d = datetime.datetime.now()
today = d.strftime("%Y-%m-%d")
#Get today's date

url = 'https://baseballsavant.mlb.com/savant-player/cole-irvin-608344?stats=statcast-r-pitching-mlb'

res = requests.get(url)
u = res.text
soup = BeautifulSoup(u, 'html.parser')
#Making request to the site and storing it in soup
```

```python
soup
```

```python
rvTable = soup.find('table', {'id': "runValues"})
```

```python
all_headings = rvTable.find_all('th')
headings = []
for i in all_headings:
    headings.append(str(i))
```

```python
all_td = rvTable.find_all('td')
table_data = []
for i in all_td:
    table_data.append(str(i))
```
