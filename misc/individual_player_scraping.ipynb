# Scraping Individual Player Stats

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
from __future__ import print_function

from datetime import date
import datetime
#imports
```

```python
batters_df = pd.read_csv('../../data/individual_batting_stats.csv')
pitchers_df = pd.read_csv('../../data/individual_pitching_stats.csv')
```

```python
url = 'https://baseballsavant.mlb.com/savant-player/cole-irvin-608344?stats=statcast-r-pitching-mlb'

res = requests.get(url)
u = res.text
soup = BeautifulSoup(u, 'html.parser')
#Making request to the site and storing it in soup
```

```python
rvTable = soup.find('table', {'id': "runValues"})
```

```python
all_headings = rvTable.find_all('th')
headings = []
for i in all_headings:
    headings.append(str(i))
```

```python
all_td = rvTable.find_all('td')
table_data = []
for i in all_td:
    table_data.append(str(i))
```

```python
headings = [i.split('</th')[0] for i in headings]
headings = [i.split('>')[1] for i in headings]
```

```python
table_data = [i.split('</')[0] for i in table_data]
table_data = [i.rsplit('>') for i in table_data]
table_data = [i[-1] for i in table_data]
#Getting data
```

```python
table_data = [table_data[x:x+18] for x in range(0, len(table_data),18)]
#Splitting it every 18 columns so it will fit into the headings column
```

```python
headings = pd.DataFrame(headings)
headings = headings.T
headings = headings.rename(columns=headings.iloc[0])
headings = headings.drop(index=0)
#changing into df then transposing and naming it so that rows are columns
```

```python
table_data = pd.DataFrame(table_data)
```

```python
table_data.columns = headings.columns
```
